```python
# importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# generar el dataset (o conjunto de datos)
dataset = make_blobs(n_samples=200,
                    centers=4,
                    n_features=2,
                    cluster_std=1.6,
                    random_state=50)
```

```python
points = dataset[0]
```

```python
# importar el k-means
from sklearn.cluster import KMeans
```

```python
# crear un objeto k-means
kmeans = KMeans(n_clusters=4)
```

```python
# ajustar el objeto k-means al dataset
kmeans.fit(points)
```

```python
plt.scatter(dataset[0][:,0], dataset[0][:,1])
# en este punto no esta dividido en clusters, solo son los datos normales
```

```python
# vamos a identificar los centroides
clusters = kmeans.cluster_centers_
```

```python
# imprimir los centroides
print(clusters)
```

```python
# reajustar los valores de los centroides hasta que haya convergencia
y_km = kmeans.fit_predict(points)
```

```python
# graficar los puntos y colorear cada cluster
plt.scatter(points[y_km == 0,0], points[y_km == 0,1], s=50, color='red')
plt.scatter(points[y_km == 1,0], points[y_km == 1,1], s=50, color='blue')
plt.scatter(points[y_km == 2,0], points[y_km == 2,1], s=50, color='yellow')
plt.scatter(points[y_km == 3,0], points[y_km == 3,1], s=50, color='green')

plt.scatter(clusters[0][0], clusters[0][1], marker='*', s=200, color='black')
plt.scatter(clusters[1][0], clusters[1][1], marker='*', s=200, color='black')
plt.scatter(clusters[2][0], clusters[2][1], marker='*', s=200, color='black')
plt.scatter(clusters[3][0], clusters[3][1], marker='*', s=200, color='black')

plt.show()
```
